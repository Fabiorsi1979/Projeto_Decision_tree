# Imports

```python
from sklearn import datasets as dt 
from sklearn import tree as tr
from matplotlib import pyplot as plt

import cv2
```

# load dataset

```python
iris = dt.load_iris ()

x = iris.data[:,2:]
y = iris.target
```

# Training

```python
# define
# max_depth = 2 tamanhpo máximo da árvore 
# min_samples_leaf = 30 número mínimo de amostras para dividir um nó, antes de fazer uma separação
# min_samples_split = 30 número mínimo de amostras para dividir um nó
# min_leaf_nodes = 30 número mínimo de folhas
#min_impurity_decrease = 30

model_tree = tr.DecisionTreeClassifier(
    max_depth=2,
    min_samples_leaf=30,
    min_samples_split=80,
    max_features=2
)
# traininig
model_tree.fit(x,y)
# predict
```

```python
tr.export_graphviz(
    model_tree,
    out_file="Tree.dot",
    feature_names=iris.feature_names[2:],
    class_names= iris.target_names,
    rounded=True,
    filled=True )
```
