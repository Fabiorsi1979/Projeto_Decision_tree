# Imports

```python
from sklearn import datasets as ds
from sklearn import model_selection as ms
from sklearn import linear_model as lm  
from matplotlib import pyplot as plt
from sklearn import metrics as mt
```

# Dataset

```python
# Generating the dataset e classes
x, y = ds.make_classification(
    n_samples=2500,
    n_classes=2,
    random_state=0)
```

# Training

```python
#training e test dataset
x_train, x_test, y_train, y_test = ms.train_test_split(x, y,
                                                       test_size=0.2,
                                                       random_state=0)
# define
model = lm.LogisticRegression(solver='lbfgs')

#train
model.fit(x_train, y_train) 

#perfomance
y_hat_test = model.predict_proba(x_test)[: , 1]
```

```python
fpr, tpr, thresholds = mt.roc_curve(y_test, y_hat_test)
```

```python
plt.plot(fpr, tpr, marker='.', label='Curva ROC')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
```

# Descobrindo o melhor valor de thresholds

```python
plt.plot(thresholds, tpr,"b--",label="TPR")
plt.plot(thresholds,1-fpr,"g-",label="FPR")
plt.vlines(0.52,0,1,colors='r',linestyles='--')
plt.hlines(0.91,0,1,colors='r',linestyles='--')
```
